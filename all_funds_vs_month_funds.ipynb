```python
import re

def extract_top_headings(md_path):
    with open(md_path, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # Match lines that start with a single '#' followed by a space
    headings = re.findall(r'(?m)^# (.+)', content)
    return headings

# Example usage
md_file = 'all_isl_merged_output.md'
headings = extract_top_headings(md_file)
print(headings)
```

```python
import re

def split_by_headings(content):
    """Split Markdown text into {heading: content} dictionary."""
    sections = {}
    parts = re.split(r'(?m)^# (.+)', content)
    # parts will look like: ['', 'Heading1', 'Content1', 'Heading2', 'Content2', ...]
    for i in range(1, len(parts), 2):
        heading = parts[i].strip()
        body = parts[i + 1].strip() if i + 1 < len(parts) else ''
        sections[heading] = body
    return sections

def remove_hashes_from_subheadings(text):
    """Remove ##, ###, etc. from subheadings but keep their text."""
    # Replace headings like '## Title' or '### Subtitle' with just 'Title'
    return re.sub(r'(?m)^#{2,}\s*', '', text).strip()

def merge_markdown(file1_path, file2_path, output_path):
    # Read both files
    with open(file1_path, 'r', encoding='utf-8') as f:
        md1 = f.read()
    with open(file2_path, 'r', encoding='utf-8') as f:
        md2 = f.read()

    # Split both into heading-content sections
    sections1 = split_by_headings(md1)
    print('----------------------SECTION_1---------------------')
    print(sections1)
    print('----------------------SECTION_1---------------------')
  
    sections2 = split_by_headings(md2)
    print('----------------------SECTION_2---------------------')
    print(sections2)
    print('----------------------SECTION_2---------------------')

    # Replace content in md1 with md2 for matching headings
    for heading in sections1:
        if heading in sections2:
            sections1[heading] = sections2[heading]
    
    # Remove subheadings (##, ###, etc.) from all sections
    for h in sections1:
        sections1[h] = remove_hashes_from_subheadings(sections1[h])
        
    # Reconstruct the updated Markdown
    merged_content = "\n\n".join(f"# {h}\n\n{sections1[h]}" for h in sections1)

    # Save the merged file
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(merged_content)

    print(f"Merged Markdown saved as: {output_path}")

md_file_1 = 'latest_modified_fmr_data/All funds latest data/June_all_fund/all_islamic_fund.md'
md_file_2 = 'latest_modified_fmr_data/single_modified_file/Islamic Fund Manager Report – September 2025.md'

# Example usage
complete_file = merge_markdown(md_file_1, md_file_2, "all_isl_merged_output.md")
```
