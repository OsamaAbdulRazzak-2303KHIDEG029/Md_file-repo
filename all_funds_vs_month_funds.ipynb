### Extrach all the single hash heading


```python
import re

def extract_top_headings(md_path):
    with open(md_path, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # Match lines that start with a single '#' followed by a space
    headings = re.findall(r'(?m)^# (.+)', content)
    return headings

```

```python
import re
from pprint import pprint

MONTH_REGEX = r'(January|February|March|April|May|June|July|August|September|October|November|December)'

def normalize_heading(h):
    """Normalize heading by removing month and year and collapsing spaces."""
    # remove "for <Month> 2025"
    h = re.sub(fr'for\s+{MONTH_REGEX}\s+\d{{4}}', '', h, flags=re.IGNORECASE)

    # remove "for <Month>"
    h = re.sub(fr'for\s+{MONTH_REGEX}', '', h, flags=re.IGNORECASE)

    # collapse multiple spaces into one
    h = re.sub(r'\s+', ' ', h)

    return h.strip()

def merge_markdown_1(file1_path, file2_path, output_path):
    with open(file1_path, 'r', encoding='utf-8') as f:
        md1 = f.read()
    with open(file2_path, 'r', encoding='utf-8') as f:
        md2 = f.read()

    sections1 = split_by_headings(md1)
    sections2 = split_by_headings(md2)

    # Create normalized lookup dictionary for md2
    norm_sections2 = {normalize_heading(h): h for h in sections2}
    # print('-------------------------')
    # print(norm_sections2)
    # print('-----------------------|')

    for h1 in list(sections1.keys()):
        norm1 = normalize_heading(h1)
        # print('===========================')
        # print(norm1)
        
        if norm1 in norm_sections2:
            print("Find matching heading")
            print(norm1)
            h2 = norm_sections2[norm1]
            print('here is h2 -----------')
            print(h2)
            print('======================')
            print(f'Replacing sections1[h1]')
            print(sections1[h1])
            print('to sections2[h2]')
            print(sections2[h2])
            sections1[h1] = sections2[h2]  # Replace September content with October content

    # Remove ##, ###, etc.
    for h in sections1:
        sections1[h] = remove_hashes_from_subheadings(sections1[h])
    
    merged_content = "\n\n".join(f"# {normalize_heading(h)}\n\n{sections1[h]}" for h in sections1)

    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(merged_content)

    print(f"Merged Markdown saved as: {output_path}")
```

```python
def append_fund_interactively(output_md, fund_heading, fund_content):
    ans = input(
        f"Do you want to add new '{normalize_heading(fund_heading)}' "
        f"in '{output_md}'? (y/n): "
    ).strip().lower()

    if ans != 'y':
        print("⏭ Skipped\n")
        return

    cleaned_content = remove_hashes_from_subheadings(fund_content)

    with open(output_md, 'a', encoding='utf-8') as f:
        f.write("\n\n")
        f.write(f"# {normalize_heading(fund_heading)}\n\n")
        f.write(cleaned_content)

    print("✅ Added\n")
```
